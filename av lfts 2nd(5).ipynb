```python
data = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test_bqCt9Pv.csv')
```

```python
train = data.copy()
y = train.loan_default
train = train.drop(['loan_default'],axis=1)

idd = test.UniqueID
```

```python
train_size = train.shape[0]
```

```python
test.head()
```

```python
#appending test and train data
train = train.append(test, ignore_index=True)
```

```python
#feature engineering

train = train.drop(['UniqueID','MobileNo_Avl_Flag'], axis=1)


category_col = ['branch_id','State_ID', 'supplier_id','manufacturer_id', 'Current_pincode_ID','Employee_code_ID',
       'Aadhar_flag', 'PAN_flag', 'VoterID_flag', 'Driving_flag',
       'Passport_flag']

for i in category_col:
    train[i] = train[i].astype('category')
```

```python
a1 = train.iloc[:,7]
b1 =[]
a2 = train.iloc[:,9]
b2_year = []
b2_month = []
a2[1][-5:-3]
```

```python
#handling date of birth
for i in range(train.shape[0]):
    b1.append(a1[i][-2:])
    b2_year.append(a2[i][-2:])
    b2_month.append(a2[i][-5:-3])
    
    
```

```python
print(len(b1))
train.shape
```

```python
train['Date.of.Birth'] = b1
train['DisbursalYear'] = b2_year
train['DisbursalMonth'] =b2_month
```

```python
train = train.drop(['DisbursalDate'],axis=1)
```

```python
train = pd.get_dummies(train, columns =['Employment.Type'])
```

```python
print(train.nunique())
print(train.dtypes)
train.head()
```

```python
train = pd.get_dummies(train, columns=['PERFORM_CNS.SCORE.DESCRIPTION'])
```

```python
a3 = train['AVERAGE.ACCT.AGE']
a4 = train['CREDIT.HISTORY.LENGTH']
b3_age = []
b4_length = []
```

```python
for i in range(train.shape[0]):
    b3_age.append(int(a3[i].split(' ')[0][:-3])*12 + int(a3[i].split(' ')[1][:-3]))
    b4_length.append(int(a4[i].split(' ')[0][:-3])*12 + int(a4[i].split(' ')[1][:-3]))
    
```

```python
train['AVERAGE.ACCT.AGE'] = b3_age
train['CREDIT.HISTORY.LENGTH'] = b4_length
```

```python
for i in train.columns:
    if(train[i].dtype=='object'):
        train[i] = train[i].astype('category')
```

```python
col = train.columns

for i in col:
    if(train[i].dtype=='int64'):
        train[i] = train[i].astype('float64')
```

```python
temp=0
category_list = []
for i in train.columns:
    if(train[i].dtype!='float64' and train[i].dtype!='uint8'):
        train[i] = train[i].astype('int32')
        category_list.append(temp)
    temp+=1
    #category_list.append(train[i].dtypes)
del(temp)
```

```python
#recovering test and train
test = train.iloc[train_size:,:]
train = train.iloc[:train_size,:]
```

```python
from sklearn.model_selection import train_test_split

train ,xval,y, yval = train_test_split(train,y, test_size=0.2, random_state=42)
```

```python
train.dtypes
```

```python
for i in ['ltv', 'asset_cost','disbursed_amount']:
    mean = train[i].mean()
    std = train[i].std()
    train[i] = (train[i] - mean)/std
    xval[i] = (xval[i]-mean)/std
```

```python
# callback
from sklearn.metrics import roc_auc_score as ras
def save_model():
    def callback(env):
        model=env.model
        if env.iteration%100==0:
            p = model.predict(xval)
            print('\n')
            print('auc score after {} itr is {}'.format(env.iteration, ras(yval,p)))
            
    #callback.before_iteration = True
    callback.after_iteration=True
    callback.order = 0
    return callback
```
